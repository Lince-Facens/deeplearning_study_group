```python
live = train.Survived.sum()
rows = train.shape[0]
percent_live = (live/rows)*100
print("a porcentagem de pessoas que viveram é: "+str(percent_live))
```

```python
dead = rows-live
percent_dead = (dead/rows)*100
print("a porcentagem de pessoas que faleceram é: "+str(percent_dead))
```
